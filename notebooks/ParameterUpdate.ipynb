```python
import xarray as xr
import numpy as np
import pandas as pd
import os
```

```python
top_dir = '/glade/work/afoster/ABoVE/parameter_files'
```

```python
pfts = ['white_spruce', 'black_spruce', 'larch', 'deciduous_trees',
        'deciduous_shrubs', 'evergreen_shrubs',
        'jack_pine', 'junper']

fates_pfts = ['needleleaf_evergreen_extratrop_tree', 'needleleaf_evergreen_extratrop_tree',
           'needleleaf_colddecid_extratrop_tree', 'broadleaf_colddecid_extratrop_tree', 
           'broadleaf_colddecid_extratrop_shrub', 'broadleaf_evergreen_extratrop_shrub',
           'needleleaf_evergreen_extratrop_tree', 'needleleaf_evergreen_extratrop_tree']

pft_df = pd.DataFrame({'NWT_PFT': pfts,
                      'PFT': fates_pfts,
                      'PFT_index': np.arange(1, len(pfts)+1)})
```

```python
old_param_file = 'fates_params_default_jan29.nc'
new_param_file = 'noresm_calib.nc'
```

```python
old_param = xr.open_dataset(os.path.join(top_dir, old_param_file))
new_param = xr.open_dataset(os.path.join(top_dir, new_param_file))
```

```python
parameters = list(old_param.data_vars)
```

```python
changed_parameters = []
for parameter in parameters:
    if parameter in list(new_param.data_vars):
        old_values = old_param[parameter]
        new_values = new_param[parameter]
        if np.array_equal(old_values, new_values):
            continue
        else:
            changed_parameters.append(parameter)
    else:
        print(parameter)
```

```python
pft_names = old_param['fates_pftname'].str.decode("utf-8").str.strip()
```

```python
i = 58
parameter = changed_parameters[i]
print(parameter)
param_df = pd.DataFrame({'PFT': pft_names,
              'old_values': old_param[parameter].values.flatten(),
              'new_values': new_param[parameter].values.flatten()})
param_df['difference'] = param_df.new_values - param_df.old_values
param_df
```
