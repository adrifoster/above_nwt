```python
import os
import pandas as pd
import numpy as np
import above_library.inventory as inv
import importlib
```

```python
inventory_dir = '/glade/work/afoster/ABoVE/obs_data/inventory_data'
inventory_files = ['NWT_Biophysical_Inventory_2015_2019_LBC.csv',
                   'NWT2022_Tree_Inventory_checkedLBC_Chevy_final_FINAL.csv',
                   'NWT2024_Tree_InventoryLBC_correctedspecies_nobalsamfir.csv'
                  ]
species_df_file = os.path.join(inventory_dir, 'NWT_species_to_PFT.csv')
site_info_file = os.path.join(inventory_dir, 'carbonstocksNWT_fulldatathrough2024_QAQCtoshare-simple.csv')

tree_eqns = pd.read_csv(os.path.join(inventory_dir, 'tree_equations.csv'))
tree_eqns['form'] = 'tree'
shrub_eqns = pd.read_csv(os.path.join(inventory_dir, 'shrub_equations.csv'))
shrub_eqns['form'] = 'shrub'
biomass_eqns = pd.concat([tree_eqns, shrub_eqns]).reset_index()
```

```python
species_df = pd.read_csv(species_df_file)
site_info = inv.clean_site_dat(site_info_file)
```

```python
site_av_dfs = []
size_count_dfs = []
height_av_dfs = []
for file in inventory_files:
    file_path = os.path.join(inventory_dir, file)
    site_avs, size_counts, height_avs = inv.get_inventory_data(file_path, species_df, site_info, biomass_eqns)
    site_av_dfs.append(site_avs)
    size_count_dfs.append(size_counts)
    height_av_dfs.append(height_avs)

site_avs = pd.concat(site_av_dfs)
size_counts = pd.concat(size_count_dfs)
height_avs = pd.concat(height_av_dfs)

site_avs["first_year"] = site_avs.groupby(["site"])["year"].transform("min")
site_avs = site_avs[site_avs.first_year == site_avs['year']]

height_avs["first_year"] = height_avs.groupby(["site"])["year"].transform("min")
height_avs = height_avs[height_avs.first_year == height_avs['year']]
```

```python
site_avs.to_csv(os.path.join('/glade/work/afoster/ABoVE/obs_data/', 'site_averages.csv'))
size_counts.to_csv(os.path.join('/glade/work/afoster/ABoVE/obs_data/', 'site_structure.csv'))
height_avs.to_csv(os.path.join('/glade/work/afoster/ABoVE/obs_data/', 'height_averages.csv'))
```

```python
site_info = site_avs[['site', 'latitude', 'longitude', 'type', 'ecoregion']].drop_duplicates()
```
